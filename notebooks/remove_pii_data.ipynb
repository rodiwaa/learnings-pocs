```python
from presidio_analyzer import AnalyzerEngine
from presidio_anonymizer import AnonymizerEngine
from presidio_anonymizer.entities import RecognizerResult, OperatorConfig

# Initialize the AnalyzerEngine
analyzer = AnalyzerEngine()
anonymizer = AnonymizerEngine()

text = """
John and his wife jane played a huge role in setting up senior accounts with account number 345-543-2345 so that others could use it for their business. they would be reachable on +1 345-322-7485 and at jane@reachme.com.
"""
scanned_results = analyzer.analyze(text=text, language='en')
redacted = anonymizer.anonymize(text=text, analyzer_results=scanned_results)
result = anonymizer.anonymize(
    text="My name is Bond, James Bond",
    analyzer_results=[
        RecognizerResult(entity_type="PERSON", start=11, end=15, score=0.8),
        RecognizerResult(entity_type="PERSON", start=17, end=27, score=0.8),
    ],
    operators={"PERSON": OperatorConfig("replace", {"new_value": "BIP"})},)

print(f"text: {text}")
print(f"redacted: {redacted.text}")

"""
Reference
https://microsoft.github.io/presidio/anonymizer/
```
